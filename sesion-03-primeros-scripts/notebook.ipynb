# Sesión 3 — Tu primer script: contar palabras de una fuente

Antes de hablar de código, hablemos de investigación. Cuando trabajamos con prensa, con archivos, con expedientes, una de las primeras preguntas que nos hacemos es bien simple: **¿qué tan larga es esta fuente? ¿qué tan rica es su vocabulario?** Una nota de cien palabras que repite tres términos no es lo mismo que una de cien palabras que usa noventa distintas. Esa diferencia —entre el total de palabras y las palabras *únicas*— es una primera medida de la densidad léxica, y los humanistas digitales la usamos para comparar fuentes, detectar redacciones formulaicas o rastrear cómo cambia el lenguaje de un diario en el tiempo.

Hoy escribimos nuestro **primer script de verdad**. No te asustes con la palabra: un script es una receta que la computadora ejecuta paso a paso. Y siempre tiene la misma forma: **entrada → proceso → salida**. Entra un archivo de texto, lo procesamos, y sale un número.

La fuente es `../datos/textos/nota_ejemplo.txt`, una nota de prensa (sintética, ficticia, inspirada en la conflictividad obrera portuaria) sobre la tensión en las plantas pesqueras. Es parte del corpus del curso.

Y una aclaración que voy a repetir hasta el cansancio: **la IA nos da el código, pero la cabeza la ponemos nosotros.** Vamos a leer cada línea, entender qué hace y validar el resultado a mano. La IA se equivoca; el rigor lo ponés vos.

## El prompt

Esto es lo que le pediríamos a la IA, en lenguaje natural, sin saber una línea de programación:

> "Escribime un script en Python que abra el archivo de texto `../datos/textos/nota_ejemplo.txt`, cuente cuántas palabras tiene en total y cuántas palabras distintas (únicas). Que muestre los dos números con un texto claro. Comentá cada paso en español para que entienda qué hace."

Fijate que el pedido es concreto: digo qué archivo, qué quiero contar y cómo quiero la salida. Un buen prompt es como una buena consigna de trabajo práctico: cuanto más clara, mejor el resultado.

## El código que la IA nos devolvió

```python
import re  # herramienta de Python para reconocer patrones en texto

# --- ENTRADA: leemos la fuente ---
# open() abre el archivo y .read() trae TODO su contenido como un solo texto.
ruta = "../datos/textos/nota_ejemplo.txt"
with open(ruta, encoding="utf-8") as archivo:
    texto = archivo.read()

# --- PROCESO: partimos el texto en palabras ---
# Pasamos todo a minúsculas para que "El" y "el" cuenten como la misma palabra.
texto = texto.lower()

# re.findall busca todas las "palabras": secuencias de letras.
# El patrón \w+ toma letras y números y descarta comas, puntos y comillas.
palabras = re.findall(r"\w+", texto)

# --- SALIDA: contamos y mostramos ---
total = len(palabras)          # cuántas palabras hay en total
unicas = len(set(palabras))    # set() descarta repetidas: nos da las distintas

print("Palabras totales:", total)
print("Palabras únicas:", unicas)
```

## Qué hace y cómo lo validamos

Leamos el script como se lee una fuente: por partes.

- **Entrada.** `open()` abre el archivo y `.read()` trae todo el contenido a la variable `texto`. Si imprimís `texto`, vas a ver la nota completa.
- **Proceso.** Primero bajamos todo a minúsculas (así "Trabajadores" y "trabajadores" no se cuentan como dos palabras distintas). Después `re.findall()` recorta el texto en palabras, dejando afuera la puntuación.
- **Salida.** `len()` cuenta cuántos elementos hay en la lista. Y `set()` se queda con una sola copia de cada palabra antes de contar: eso nos da las únicas.

¿Y cómo sabemos que el número está bien? **Validando a mano.** La nota es corta: podés contar las palabras del primer párrafo vos mismo y ver si el orden de magnitud cierra. Si el script dijera "5 palabras totales", sabrías que algo se rompió. Esa desconfianza sana es lo que separa a quien investiga de quien copia y pega.

```python
# Mostramos las primeras 15 palabras detectadas para chequear que tengan sentido
print(palabras[:15])
```

Si ahí ves cosas como `'crece'`, `'la'`, `'tensión'`, vas por buen camino: el script está leyendo la fuente de verdad.

## Modificamos el script: contar UNA palabra concreta

Acá viene lo importante de la sesión: **perder el miedo a tocar el código.** Ya tenemos las palabras separadas; ahora queremos saber cuántas veces aparece un término que nos interesa como investigadores. En una nota sobre conflicto laboral, palabras como "sindicato" o "paritarias" son señales. Contarlas es un primer gesto de análisis de discurso.

Este sería el pedido a la IA:

> "Modificá el script anterior para que, además, me diga cuántas veces aparece la palabra «sindicato» en la nota."

Y el cambio es mínimo. Mirá:

```python
# Elegí acá la palabra que querés rastrear. ¡Cambiá este valor y volvé a correr!
palabra_buscada = "sindicato"

# .count() cuenta cuántas veces ese valor aparece en la lista que ya teníamos.
veces = palabras.count(palabra_buscada)

print(f"La palabra '{palabra_buscada}' aparece {veces} vez/veces.")
```

¿Ves la línea `palabra_buscada = "sindicato"`? Eso es una **variable**: una cajita con un nombre donde guardamos un valor. El resto del código no sabe ni le importa qué palabra pusiste; trabaja con lo que haya en la caja. Por eso cambiar la palabra es tan fácil: tocás un solo lugar.

**Validación a ojo:** abrí `nota_ejemplo.txt` y buscá "sindicato" con Ctrl+F. Contá las coincidencias. ¿Coincide con lo que dijo el script? Si sí, podés confiar. Si no, antes de creerle al programa, preguntate por qué: ¿escribiste bien la palabra? ¿hay mayúsculas que no convertimos? Esa pregunta vale más que el número.

## Qué hacer ante un error

Tarde o temprano algo va a fallar, y está perfecto. Un error no es un reto: es la computadora pidiéndote precisión. Probemos romperlo a propósito para perderle el miedo.

Si escribís mal la ruta del archivo, Python te va a frenar con un `FileNotFoundError`. Ese mensaje no es una catástrofe: dice, básicamente, "no encuentro el archivo". La traducción práctica es: revisá el nombre y la ruta.

**Cuando algo falle, copiá el mensaje de error y pegáselo a la IA tal cual**, con un "me da este error, ¿qué significa y cómo lo arreglo?". Es la forma más rápida de aprender. La celda de abajo está pensada para fallar a propósito: corréla y leé el mensaje en rojo con calma.

```python
# A propósito ponemos una ruta que NO existe para ver el error.
# (Esta celda DEBE fallar: es parte de la lección.)
with open("../datos/textos/archivo_que_no_existe.txt", encoding="utf-8") as f:
    contenido = f.read()
```

Y si querés que el script no explote cuando el archivo no está, sino que avise con cortesía, se lo podés pedir a la IA. Una versión así:

```python
import os

ruta = "../datos/textos/nota_ejemplo.txt"

# os.path.exists() devuelve True si el archivo está, False si no.
if os.path.exists(ruta):
    with open(ruta, encoding="utf-8") as archivo:
        texto = archivo.read()
    print("Archivo leído. Caracteres:", len(texto))
else:
    print("No encontré el archivo en:", ruta)
    print("Revisá que el nombre y la carpeta sean correctos.")
```
