## 4. Limpiar lo que haga falta

Los datos reales vienen sucios: espacios de más, mayúsculas inconsistentes, restos de HTML, fechas con formato raro, columnas que sobran. Limpiar es parte del trabajo histórico, no un trámite previo. Adaptá estos pasos a la suciedad concreta de TU fuente.

```python
# TODO: limpiá según lo que tengan TUS datos. Algunos ejemplos típicos:

datos_limpios = datos.copy()

# 1) Sacar espacios sobrantes al principio y al final de los textos
for col in datos_limpios.select_dtypes(include="object").columns:
    datos_limpios[col] = datos_limpios[col].str.strip()

# Otros arreglos frecuentes (descomentá y adaptá los que necesites):
# datos_limpios["texto"] = datos_limpios["texto"].str.replace(r"<[^>]+>", "", regex=True)  # restos de HTML
# datos_limpios["texto"] = datos_limpios["texto"].str.replace("&nbsp;", " ", regex=False)  # entidades HTML
# datos_limpios = datos_limpios.dropna(subset=["texto"])                                    # filas vacías

# Validá: ¿bajó o cambió algo? Compará el "antes" y el "después".
print("Filas antes:", len(datos))
print("Filas después:", len(datos_limpios))
datos_limpios.head()
```

## 5. Analizar

Acá respondés tu pregunta. El análisis correcto depende de qué le preguntás a tu fuente: contar, agrupar, comparar, medir frecuencias. Abajo dejé dos andamios habituales; quedate con el que sirva a TU pregunta y borrá el otro.

```python
# --- ANDAMIO A: contar / agrupar por una columna (para datos en tabla) ---
# TODO: cambiá "seccion" por la columna que querés resumir.

resumen = datos_limpios["seccion"].value_counts()   # ¿cuántos registros hay por cada categoría?
resumen
```

```python
# --- ANDAMIO B: contar palabras (para fuentes textuales) ---
# TODO: cambiá "texto" por tu columna de texto. Ajustá tus propias "stopwords".

# Palabras vacías que no aportan sentido. AGREGÁ las que sobren en tu corpus.
vacias = {"de", "la", "el", "que", "en", "y", "a", "los", "las", "un", "una",
          "del", "por", "con", "para", "se", "su", "al", "lo", "como", "más"}

# Junto todo el texto, lo paso a minúsculas y lo parto en palabras
todo = " ".join(datos_limpios["texto"].astype(str)).lower()
palabras = re.findall(r"[a-záéíóúñ]+", todo)
palabras = [p for p in palabras if p not in vacias]   # saco las vacías

frecuencias = Counter(palabras)

# Las 15 palabras más frecuentes: ¿asoma el relato de tu fuente en los números?
frecuencias.most_common(15)
```

## 6. Visualizar

Un gráfico no es decoración: es un argumento. Tiene que dejar ver de un vistazo eso que descubriste en el análisis. Adaptá el código a TU resultado (cambiá los datos de origen, los ejes y, sobre todo, el título y las etiquetas, que son las que comunican).

```python
# TODO: cambiá "frecuencias" por tu resultado y ajustá las etiquetas.

top = frecuencias.most_common(10)            # me quedo con las 10 más frecuentes
etiquetas = [palabra for palabra, _ in top]
valores = [n for _, n in top]

fig, ax = plt.subplots(figsize=(7, 4.5))
ax.barh(etiquetas, valores, color="steelblue")
ax.invert_yaxis()                            # la más frecuente arriba
ax.set_title("TODO: poné un título que diga qué se ve")
ax.set_xlabel("Frecuencia")
fig.text(0.01, -0.04, "Fuente: TODO — citá tus datos", fontsize=8)
plt.tight_layout()
plt.show()
```
